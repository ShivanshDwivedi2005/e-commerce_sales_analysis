```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

the file we are going to work is .zip file so first we need to change it in xlsx file

```python
import zipfile

# Path to the ZIP file
zip_path = "Superstore_USA.xlsx.zip"
extract_path = "./"  # Directory to extract files to

# Extract the ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("File extracted successfully!")
```

```python
import pandas as pd

# Path to the extracted Excel file
excel_file_path = "Superstore_USA.xlsx"

# Load the data into a DataFrame
data = pd.read_excel(excel_file_path)

# Display the first few rows
print(data.head())
```

```python
data.isnull().sum()
```

there are 72 rows in which product base margin is missing now i need to fill that cells with mean 

```python
data["Product Base Margin"] = data["Product Base Margin"].fillna(data["Product Base Margin"].mean())
```

```python
data.isnull().sum()
```

```python
ax = sns.countplot(x = 'Order Priority',data = data,hue = 'Discount')
plt.title("order Priority count")
plt.show()
data['Order Priority'].value_counts()
```

```python
data['Order Priority'].unique()
```

the above data show that theri are two critical because of an extra space in second critical now we need to replace 'Critical ' with 'Critical'

```python
data = data.replace("Critical ","Critical")
data['Order Priority'].unique()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize = (22,8))
ax = sns.countplot(x = 'Unit Price', data = df)
plt.title('total sales count per unit price')
plt.show()
```

```python
from matplotlib import style
import matplotlib.pyplot as plt
style.use('ggplot')
unit_price_counts = df['Unit Price'].value_counts()
plt.scatter(unit_price_counts.index, unit_price_counts.values)
plt.xlabel('Unit Price')
plt.ylabel('Count')
plt.title('Unit Price vs Count')
plt.show()
```

* most of the projects are of less price ranges within 1000

```python
style.use('ggplot')
unit_price_counts = df['Shipping Cost'].value_counts()
plt.scatter(unit_price_counts.index, unit_price_counts.values)
plt.xlabel('Shipping Cost')
plt.ylabel('Count')
plt.title('Unit Price vs Count')
plt.show()
```

* generally people prefer to order the products with less shipping cost

```python
plt.figure(figsize=(16,8))
ax = sns.countplot(x = 'Product Category', data = df,hue = 'Product Sub-Category')
plt.title('Product Category count')
plt.show()
```
